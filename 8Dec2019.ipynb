#### Apply linear regression on LifeExpectancy.csv data
Use the given dataset which includes Country's GDP,Fertility rate and Population.

Your goal is to predict the life expectancy in a given region by considering target variable as "life" 
and feature variable as "fertility".

1)Apply Linear Regression to the given dataset without splitting dataset
2)Check the R squared score
3)Split the dataset into training and testing sets.
4)Check the R squared score
5)Finally take any value for fertility and predict the life.

```python
import pandas as pd
data=pd.read_csv("Lifeexpectancy.csv")
data.head(2)
```

```python
x=data["fertility"] .values#input feature
y=data["life"].values    #target
x=x.reshape((-1,1))
```

```python
#Applying the model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
LR=reg.fit(x,y)
```

```python
#R2 score
print(reg.score(x,y))
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
LR=reg.fit(x_train,y_train)
```

```python
#r2 score after splitting
print(reg.score(x_test,y_test))
```

```python
reg.predict([[2.73]])
```

## Classification
* Logistic regression(binomial,multinomial)
* binomial(0,1 or yes or no)
* multinomial(0,1,2 and more)

```python
#binomial
df=pd.read_csv("diabetes.csv")
df.head(2)
```

```python
x=df.iloc[:,0:8].values#input features
y=df["Outcome"].values  #target

```

```python
#apply the model
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
LR=reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
y_pred
```

```python
reg.predict([[1,85,66,29,0,26.6,0.351,31]])
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn import datasets
```

```python
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_data["target"]=iris.target
iris_data
```

```python
x=iris_data.iloc[:,0:4].values #input features
y=iris_data["target"].values   #target
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(multi_class="multinomial",solver='lbfgs')
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
LR=reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
