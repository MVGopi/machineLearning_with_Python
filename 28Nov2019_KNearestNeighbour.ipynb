```python
import pandas as pd
shirts_df = pd.read_csv("shirts.csv")
shirts_df
```

```python
height = shirts_df['Height(cms)'].values
weight = shirts_df['Weight(inKgs)'].values
#import numpy as np
import math
distance = []
x2 = 161
y2 = 61
```

```python
for i in range(len(height)):
   distance.append((math.sqrt((x2-height[i])**2+(y2-weight[i])**2)))
```

```python
shirts_df['distance'] = distance
```

```python
shirts_df
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(shirts_df['T-shirt'].values)
target = le.transform(shirts_df['T-shirt'].values)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(shirts_df.iloc[:, 0:2], target, test_size = 0.25, random_state = 2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors = 5)
knn_clf.fit(X_train, Y_train)
```

```python
y_predict = knn_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_predict)
```
